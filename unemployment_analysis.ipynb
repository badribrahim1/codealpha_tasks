```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\Users\badre\OneDrive\Desktop\Unemployment in India\Unemployment in India.csv")
print(df.head())
```

```python
print(df.head())
print(df.shape)
print(df.columns)
```

```python
print(df.isnull().sum())
df = df.dropna()
```

```python
if "Date" in df.columns:
    df["Date"] = pd.to_datetime(df["Date"])
```

```python
print(df.columns)
```

```python
print(df.describe())
```

```python
if "Region" in df.columns:
    print(df.groupby("Region").mean(numeric_only=True))
```

```python
sns.histplot(df.iloc[:, -1], kde=True)
plt.title("Data Distribution")
plt.show()
```

```python
num_col = df.select_dtypes(include=np.number).columns[0]

if "Date" in df.columns:
    sns.lineplot(data=df, x="Date", y=num_col)
    plt.title("Unemployment Trend Over Time")
    plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
